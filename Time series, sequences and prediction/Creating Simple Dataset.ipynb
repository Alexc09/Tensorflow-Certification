```
import tensorflow as tf
```

```
#Create a dataset with values from 0 to 9
ds = tf.data.Dataset.range(10)

for i in ds:
  print(i.numpy())
```

```
ds = tf.data.Dataset.range(10)
#Create a window of size 5. Each window has a max of 5 elements and will increase the numbers by 1 (Since shift = 1)
ds = ds.window(5,shift=1)
for window in ds:
  for val in window:
    print(val.numpy(), end=' ')
  print()
```

```
ds = tf.data.Dataset.range(10)
#Drop_remainder so windows will be dropped if its size is less than 5
ds = ds.window(5,shift=1, drop_remainder=True)
for window in ds:
  for val in window:
    print(val.numpy(), end=' ')
  print()
```

```
ds = tf.data.Dataset.range(10)
ds = ds.window(5,shift=1, drop_remainder=True)
#Write a function that flattens the ds and batches it into groups of 5 (So you can see it as arrays containing 5 elements each)
#You use flat_map to ensure the order of data is the same. It flattens all elements so the ds will be a 1D array
ds = ds.flat_map(lambda window:window.batch(5))
for window in ds:
  print(window.numpy())
```

```
ds = tf.data.Dataset.range(10)
ds = ds.window(5,shift=1, drop_remainder=True)
ds = ds.flat_map(lambda window:window.batch(5))
#Create a function that takes the data and returns 2 sets of values (one for X and one for Y)
ds = ds.map(lambda data: (data[:-1], data[-1:]))
for x,y in ds:
  print(x.numpy(), y.numpy())
```

```
ds = tf.data.Dataset.range(10)
ds = ds.window(5,shift=1, drop_remainder=True)
ds = ds.flat_map(lambda window:window.batch(5))
ds = ds.map(lambda data: (data[:-1], data[-1:]))
#Shuffle the dataset using buffer size of 10
ds = ds.shuffle(buffer_size = 10)
for x,y in ds:
  print(x.numpy(), y.numpy())
```

```
ds = tf.data.Dataset.range(10)
ds = ds.window(5,shift=1, drop_remainder=True)
ds = ds.flat_map(lambda window:window.batch(5))
ds = ds.map(lambda data: (data[:-1], data[-1:]))
ds = ds.shuffle(buffer_size = 10)
#Batch the dataset into batches of 2 (One batch will have 2 Xs and the 2 corresponding Ys)
ds = ds.batch(2).prefetch(1)
for x,y in ds:
  print(x.numpy(), y.numpy())
```

```
series = [20.7, 17.9, 18.8, 14.6, 15.8, 15.8, 15.8, 17.4, 21.8, 20.]
```

```
new = tf.expand_dims(series, axis=-1)
ds = tf.data.Dataset.from_tensor_slices(new)
for i in ds:
  print(i.numpy())
```

```
ds = tf.data.Dataset.from_tensor_slices(new)
ds = ds.window(5, shift=1, drop_remainder=True)
ds = ds.flat_map(lambda window:window.batch(5))
for window in ds:
  print(window.numpy())
```

```
new = tf.expand_dims(series, axis=-1)
ds = tf.data.Dataset.from_tensor_slices(new)
ds = ds.window(5, shift=1, drop_remainder=True)
ds = ds.flat_map(lambda window:window.batch(5))
ds = ds.map(lambda x: (x[:-1], x[-1:]))
ds = ds.batch(3).prefetch(1)
for x in ds:
  print(x)
  print()
```
