```
tm = open('sonnets.txt','r').read()
corpus = tm.lower().split('\n')
```

```
tokenizer = Tokenizer()
tokenizer.fit_on_texts(corpus)
word_idx = tokenizer.word_index
```

```
print(word_idx)
```

```
rev_word_idx = {word_idx[i]:i for i in list(word_idx)}
```

```
sen = 'This is a great day'
next_words  = 10
mlen = 16

for _ in range(next_words):
  enc = tokenizer.texts_to_sequences([sen])[0]
  padded = pad_sequences([enc], maxlen=mlen-1, padding='pre', truncating='pre')
  pred_class = model.predict_classes(padded)
  pred_word = ''
  for i in rev_word_idx:
    if i == pred_class:
      pred_word = rev_word_idx[i]
      break
  sen = f'{sen} {pred_word}'

print(sen)
```
